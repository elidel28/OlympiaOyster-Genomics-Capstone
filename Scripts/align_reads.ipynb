```python
import subprocess
import os

ref = "/mnt/jupiter/home/edelga20/annotation_jasmine-the2899-mb-hirise-ahy4t__02-20-2022__hic_output.fasta"
out_dir = "results"
os.makedirs(out_dir, exist_ok=True)

# Absolute base path to your data
base_path = "/mnt/jupiter/home/edelga20/"

# ⬇️ Your custom list of FASTQ files

#"johnsonlab/lcwg_Olympia/14KP_TTAACGGTGT-ACGGTCAGGA_L004_R1_001.fastq.gz",
#"johnsonlab/lcwg_Olympia/14KP_TTAACGGTGT-ACGGTCAGGA_L004_R2_001.fastq.gz",

fastq_files = [
    "johnson/johnsonlab/lcwg_Olympia/69SMF_TCCGGACTAG-ATTGTTCGTC_L004_R1_001.fastq.gz", 
    "johnson/johnsonlab/lcwg_Olympia/69SMF_TCCGGACTAG-ATTGTTCGTC_L004_R2_001.fastq.gz", 
    "johnson/johnsonlab/lcwg_Olympia/70SMF_GCACCACCAA-ATCAGTACCA_L004_R1_001.fastq.gz", 
    "johnson/johnsonlab/lcwg_Olympia/70SMF_GCACCACCAA-ATCAGTACCA_L004_R2_001.fastq.gz", 
    "johnson/johnsonlab/lcwg_Olympia/8KP_ATAGCGGAAT-TACTCTGCGC_L004_R1_001.fastq.gz", 
    "johnson/johnsonlab/lcwg_Olympia/8KP_ATAGCGGAAT-TACTCTGCGC_L004_R2_001.fastq.gz", 
]

fastq_files_full = [os.path.join(base_path, f) for f in fastq_files]

# ⬇️ Group R1/R2 pairs by sample prefix
paired_samples = {}
for f in fastq_files_full:
    base = os.path.basename(f)
    sample_id = base.split("_")[0]
    if "_R1_" in f:
        paired_samples.setdefault(sample_id, {})["R1"] = f
    elif "_R2_" in f:
        paired_samples.setdefault(sample_id, {})["R2"] = f

# ⬇️ Align with BWA for each paired file
for sample, files in paired_samples.items():
    r1 = files.get("R1")
    r2 = files.get("R2")
    if r1 and r2:
        output_sam = os.path.join(out_dir, f"{sample}.sam")
        print(f"🔄 Aligning {sample} → {output_sam}")
        cmd = ["/bin/bwa", "mem", ref, "-t", "10", r1, r2]
        with open(output_sam, "w") as out_f:
            subprocess.run(cmd, stdout=out_f)
    else:
        print(f"⚠️  Incomplete pair for {sample} — skipping.")

result = subprocess.run(cmd, stdout=out_f, stderr=subprocess.PIPE)
if result.returncode != 0:
    print(f"Error aligning {sample}: {result.stderr.decode()}")

```

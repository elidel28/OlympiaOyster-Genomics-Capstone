```python
"""
Make this more efficient by using bwa-mem2 instead of bwa mem, and also parallelize the process to process multiple files at once
"""

# BWA alignment with Read Groups
import subprocess
import os
import gzip

# Reference sequence
ref = "/mnt/jupiter/home/chdahl/annotation_jasmine-the2899-mb-hirise-ahy4t__02-20-2022__hic_output.fasta"

# Out directory (where created files are sent)
out_dir = "test_results"

# Makes the directory if there isn't one already with that name
os.makedirs(out_dir, exist_ok=True)

# Threads allocated to the tool
threads = "2" #change to use however many you need

# Absolute base path to your data
base_path = "/mnt/jupiter/home/chdahl/" #alter to relevant base path

# ‚¨áÔ∏è Your custom list of FASTQ files
fastq_files = [
    "johnsonlab/lcwg_Olympia/10KP_GTTCGGAGTT-TCAGAAGGCG_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/10KP_GTTCGGAGTT-TCAGAAGGCG_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/14KP_TTAACGGTGT-ACGGTCAGGA_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/14KP_TTAACGGTGT-ACGGTCAGGA_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/15AZ_ACGTTCCTTA-CACAGCGGTC_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/15AZ_ACGTTCCTTA-CACAGCGGTC_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/28KP_TTGTATCAGG-TGGCCTCTGT_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/28KP_TTGTATCAGG-TGGCCTCTGT_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/31SMW_GTGACGGAGC-TGGACCGCCA_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/31SMW_GTGACGGAGC-TGGACCGCCA_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/32SMW_CCACGCTGAA-CTGAGGAATA_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/32SMW_CCACGCTGAA-CTGAGGAATA_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/34SMW_TATGACAATC-ATACAAGCAA_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/34SMW_TATGACAATC-ATACAAGCAA_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/35SMW_CACAGCGGTC-CAATAGGAAT_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/35SMW_CACAGCGGTC-CAATAGGAAT_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/36SMW_TTAACCTTCG-TGTCTGGCCT_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/36SMW_TTAACCTTCG-TGTCTGGCCT_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/38SMW_AATTCCATCT-GTAACTGAAG_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/38SMW_AATTCCATCT-GTAACTGAAG_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/39SMW_ATGTTGTTGG-TCATAGATTG_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/39SMW_ATGTTGTTGG-TCATAGATTG_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/46SMW_AGCCTATGAT-GTAGGTGGTG_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/46SMW_AGCCTATGAT-GTAGGTGGTG_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/52SMW_ACTTGTTATC-TCTAGGCGCG_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/52SMW_ACTTGTTATC-TCTAGGCGCG_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/55SMW_CATATGCGAT-CCGTTCAAGG_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/55SMW_CATATGCGAT-CCGTTCAAGG_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/57SMW_CGTGTACCAG-AACTTATCCT_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/57SMW_CGTGTACCAG-AACTTATCCT_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/69SMF_TCCGGACTAG-ATTGTTCGTC_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/69SMF_TCCGGACTAG-ATTGTTCGTC_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/70SMF_GCACCACCAA-ATCAGTACCA_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/70SMF_GCACCACCAA-ATCAGTACCA_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/8KP_ATAGCGGAAT-TACTCTGCGC_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/8KP_ATAGCGGAAT-TACTCTGCGC_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/9KP_GCAATATTCA-AATTGGCGCC_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/9KP_GCAATATTCA-AATTGGCGCC_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/MOR021_TTACGCACCT-CAACGAGAGC_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/MOR021_TTACGCACCT-CAACGAGAGC_L004_R2_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/MOR023_CGACAAGGAT-AGGTGGCCTT_L004_R1_001.fastq.gz",
    "johnsonlab/lcwg_Olympia/MOR023_CGACAAGGAT-AGGTGGCCTT_L004_R2_001.fastq.gz"
] #put your relevant file paths

fastq_files_full = [os.path.join(base_path, f) for f in fastq_files]

def extract_readgroup_info(r1_path, sample_id):
    with gzip.open(r1_path, "rt") as f:
        header = f.readline().strip()
    header_fields = header.split(":")
    
    # Extract instrument, PU from header
    instrument = header_fields[0].replace("@", "")
    flowcell = header_fields[2]
    lane = header_fields[3]
    pu = f"{flowcell}.{lane}"

    # Construct @RG string
    rg_dict = {
        'ID': instrument,
        'SM': sample_id,
        'PL': 'ILLUMINA',
        'LB': 'LCWG_Olympia', #The different libraries for other samples are in different file paths, all of our current test data is from this folder
        'PU': pu
    }

    # Format and escape tabs
    rg_string = "@RG\\t" + "\\t".join([f"{k}:{v}" for k, v in rg_dict.items()])
    return rg_string

# Group R1/R2 pairs by sample prefix
paired_samples = {}
for f in fastq_files_full:
    base = os.path.basename(f)
    sample_id = base.split("_")[0]
    if "_R1_" in f:
        paired_samples.setdefault(sample_id, {})["R1"] = f
    elif "_R2_" in f:
        paired_samples.setdefault(sample_id, {})["R2"] = f

# Align with BWA using extracted RG info
for sample, files in paired_samples.items():
    r1 = files.get("R1")
    r2 = files.get("R2")
    if r1 and r2:
        #test again
        #output_sam = os.path.join(out_dir, f"{sample}_.sam")
        output_sam = os.path.join(out_dir, f"{sample}_test.sam")
        print(f"üîÑ Aligning {sample} ‚Üí {output_sam}")

        # Extract RG info from the R1 fastq

        #removing this temporarily
        #rg = extract_read_group_info(r1)
        rg = extract_readgroup_info(r1, sample)
        if not rg:
            print(f"‚ö†Ô∏è  Could not extract RG info for {sample} ‚Äî skipping.")
            continue

        # Construct RG string for BWA

        #old string, testing the new one
        rg_string = extract_readgroup_info(r1, sample)

        # BWA mem command with x threads and read group
        cmd = [
            "/mnt/jupiter/home/chdahl/miniconda3/bin/bwa", "mem",
            "-t", threads,
            "-R", rg_string, #this is the read group header
            ref,  # your reference genome
            r1, r2
        ]

        with open(output_sam, "w") as out_f:
            subprocess.run(cmd, stdout=out_f)
    else:
        print(f"‚ö†Ô∏è  Incomplete pair for {sample} ‚Äî skipping.")
```
